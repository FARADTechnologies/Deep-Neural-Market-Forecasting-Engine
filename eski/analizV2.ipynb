```python
import pandas as pd


df = pd.read_csv("final_sales_data_CLEANED.csv", 
                 encoding="utf-8-sig", 
                 dtype={'all_pictures': str})
```

```python
# 1. Kaç tane dolu (NaN olmayan) deed_number değeri var?
dolu_sayisi = df['deed_number'].count()

# 2. Bu değerlerin listesi (İlk 10 tanesini görelim)
degerler = df['deed_number'].unique()

print(f"Toplam {len(df)} satırdan {dolu_sayisi} tanesinde deed_number verisi dolu.")
print(f"Sütundaki benzersiz değerlerin sayısı: {len(degerler)}")
print("\nÖrnek deed_number değerleri:")
print(degerler[:10]) # İlk 10 tanesini listeler
```

```python


# 1. Her bir deed_number'ın kaç kez geçtiğini hesapla
deed_counts = df['deed_number'].value_counts()

# 2. Sadece 1'den fazla geçenleri (yani tekrar edenleri) filtrele
duplicates = deed_counts[deed_counts > 1]

# 3. Sonuçları yazdır
print(f"Tekrar eden benzersiz 'deed_number' sayısı: {len(duplicates)}")
print(f"Toplam tekrar eden satır sayısı: {df.duplicated(subset=['deed_number']).sum()}")
print("\nEn çok tekrar eden ilk 10 deed_number ve sayıları:")
print(duplicates.head(10))



df.head(10)
```

```python

df.drop(columns=['createdAt', 'updatedAt', 'link', 'all_pictures', 'purpose'], inplace=True, errors='ignore')

pd.options.display.max_columns = None

df.head(10)
```

```python

cat_cols = [
    'id', 'name', 'location', 'lat', 'long', 'building_no', 'Amenities', 
    'furnished', 'completion_status', 'reactivatedAt', 'city', 'region', 
    'district', 'link', 'building_type', 'all_pictures', 'purpose', 'type', 
    'postal_code', 'additional_no', 'isVerified', 'listing_age', 'listing_face', 
    'deed_number', 'hasProject', 'is_listing_constrained', 'is_listing_pawned', 
    'agency_name', 'residence_type'
]


num_cols = [col for col in df.columns if col not in cat_cols and col not in ['index', 'createdAt', 'updatedAt']]

print(f"Kategorik Sütun Sayısı: {len(cat_cols)}")
print(f"Sayısal Sütun Sayısı: {len(num_cols)}")
```

```python
# Kategorik sütunları metin (string) yap
df[cat_cols] = df[cat_cols].astype(str)

# Sayısal sütunları sayıya (numeric) çevir
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("Dönüşüm tamamlandı!")
```

```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Sayısal olanların özet tablosu (ortalama, max vb.)
display(df[num_cols].describe())

# Kategorik olanların ilk 5 satırı
display(df[cat_cols].head())
```

```python

num_analysis = []
for col in num_cols:
    num_analysis.append({
        'Column': col,
        'Null Values': df[col].isna().sum(),
        'Min Value': df[col].min(),
        'Max Value': df[col].max()
    })

num_summary_df = pd.DataFrame(num_analysis)

cat_analysis = []
for col in cat_cols:
    cat_analysis.append({
        'Column': col,
        'null': df[col].replace('nan', pd.NA).isna().sum(), 
        'Unique': df[col].nunique()
    })

cat_summary_df = pd.DataFrame(cat_analysis)


print(f"Toplam Sayısal Sütun Sayısı: {len(num_cols)}")
display(num_summary_df)

print(f"\nToplam Kategorik Sütun Sayısı: {len(cat_cols)}")
display(cat_summary_df)
```

```python

num_unique_analysis = []

for col in num_cols:
    num_unique_analysis.append({
        'Sütun Adı': col,
        'uniqe': df[col].nunique(),
        'sample': df[col].unique()[:5].tolist() 
    })

num_unique_df = pd.DataFrame(num_unique_analysis)
display(num_unique_df)
```
